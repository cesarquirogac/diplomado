# Manual

```python
import glob
```

```python
folder_path = 'manual/'
```

```python
xlsx_files = glob.glob(folder_path + '*.xlsx')
```

```python
combined_df = pd.DataFrame()
```

```python
for file in xlsx_files:
    df = pd.read_excel(file, header=None)
    combined_df = pd.concat([combined_df, df], ignore_index=True)
```

```python
combined_df
```

```python
new_df = pd.DataFrame()
for index, row in combined_df.iterrows():
    id = str(row[0])
    columns_value = row[5:].dropna().values
    temp_df = pd.DataFrame({'ID': id, 'Comment': columns_value})
    new_df  = pd.concat([new_df, temp_df], ignore_index=True)
```

# JSON

```python
json_dir = 'json'
```

```python
json_df = pd.DataFrame()
for filename in os.listdir(json_dir):
    if filename.endswith('.json'):
        id = str(filename.replace(".json", ""))
        file_path = os.path.join(json_dir, filename)
        with open(file_path, 'r') as file:
            data_list = json.load(file)
        content_values = [entry.get('content') for entry in data_list]
        temp_df = pd.DataFrame({'ID': id, 'Comment': content_values})
        json_df  = pd.concat([json_df, temp_df], ignore_index=True)
```

```python
json_df.head()
```

```python
json_df.shape
```

# Export Comments

```python
xlsx_dir = 'exportcomments'
```

```python
export_df = pd.DataFrame()
for filename in os.listdir(xlsx_dir):
    if filename.endswith('.xlsx'):
        id = str(filename.replace(".xlsx", ""))
        file_path = os.path.join(xlsx_dir, filename)
        temp_df = pd.read_excel(file_path, header=None, skiprows=7)
        content_values = temp_df[7].tolist()
        temp_df = pd.DataFrame({'ID': id, 'Comment': content_values})
        export_df = pd.concat([export_df, temp_df], ignore_index=True)
```

```python
export_df.head()
```

```python
export_df.shape
```

# Merge all comments
